# Clustering using K-Means with Titanic Dataset

First, let's import necessary libraries.

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
import seaborn as sns
```

Then let's read train set and test set.

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

Let's examine embarked column.<br>
The number of passengers getting from S port is higher than others.

```python
sns.countplot(train_data['Embarked'], hue=train_data['Survived'], palette='BuPu')
```

We fill in the lost values according to the most preferred port. So with S.

```python
train_data['Embarked'] = train_data['Embarked'].fillna("S")
train_data.info()
```

```python
train_data.groupby(['Embarked']).mean()
```

We see that people boarding at port C pay the highest fare, with the highest survival rate. The most crowded port S but survival probability of C have more than others.

Another column with missing values is age column.<br>
We find the average age value and replace it with missing data.<br>
Since it will be a float value, we finally rounded the value using round() function.

```python
age_mean = train_data['Age'].mean()
train_data['Age'] = train_data['Age'].replace(np.nan, age_mean)
train_data['Age'] = train_data['Age'].round()
```

While doing data analysis, every column may not be necessary for us.<br>
Also, not removing the columns that are not required for analysis may distort our results.<br>
There were many missing values in the cabin column. We cannot fill in so many missing values. This changes our results considerably. If we do, we can skew the result.

```python
d = ['PassengerId','Name','Ticket','Cabin']
train_data = train_data.drop(d, axis=1)
train_data.head()
```

```python
train_data.isnull().sum()
```

## One-Hot Encoding

This is one-hot encoding part.<br>
Sex and embarked columns consist of categorical variables.<br>
To convert categorical variables into numerical variables, we will use get_dummies() function of Pandas for this case.

```python
embarked = pd.get_dummies(train_data['Embarked'], drop_first = True)
train_data = pd.concat([train_data, embarked], axis = 1)
train_data.drop(['Embarked'] , axis = 1 , inplace = True) #now this can be deleted
train_data.head()
```

```python
sex = pd.get_dummies(train_data['Sex'], drop_first = True)
train_data = pd.concat([train_data, sex], axis = 1)
train_data.drop(['Sex'] , axis = 1 , inplace = True) #now this can be deleted
train_data.head()
```

We can convert the SibSp and Parch columns into one column. Thus, we can get a better data set.<br>
I collect the SibSp and Parch column as TravelAlone in one column.<br>
The SibSp and Parch column meant things like spouse, parent, child accompanying the journey.

```python
train_data['TravelAlone'] = np.where((train_data['SibSp'] + train_data['Parch'])> 0, 0, 1)
```

```python
train_data.head()
```

```python
train_data = train_data.drop(['SibSp','Parch'], axis=1)
train_data.head()
```

Let's do the same for the test set.

```python
test_data.isnull().sum()
```

```python
test_data['Age'] = train_data['Age'].replace(np.nan, age_mean)
test_data['Age'] = test_data['Age'].round()
```

```python
fare_median = test_data['Fare'].median(skipna=True)
test_data['Fare'] = test_data['Fare'].replace(np.nan, fare_median)
```

We found that there was too much missing value in the cabin column. If we complete this missing data, we probably distort our result.<br>
Therefore, we are removing this column.

```python
test_data = test_data.drop('Cabin', axis=1)
drop_t = ['PassengerId','Name','Ticket']
test_data = test_data.drop(drop_t, axis=1)
test_data.head()
```

```python
test_data.isnull().sum()
```

```python
test_data.head()
```

```python
embarked = pd.get_dummies(test_data['Embarked'], drop_first = True) #one-hot encoding
test_data = pd.concat([test_data, embarked], axis = 1)
test_data.drop(['Embarked'] , axis = 1 , inplace = True)
test_data.head()
```

```python
sex = pd.get_dummies(test_data['Sex'], drop_first = True) #one-hot encoding
test_data = pd.concat([test_data, sex], axis = 1)
test_data.drop(['Sex'] , axis = 1 , inplace = True)
test_data.head()
```

```python
test_data.head()
```

```python
test_data['TravelAlone'] = np.where((test_data['SibSp']+test_data['Parch'])>0, 0, 1)
test_data = test_data.drop(['SibSp','Parch'], axis=1)
```

```python
#X = train_data.drop('Survived', axis=1) 
#y = train_data['Survived'] 
X = np.array(train_data.drop(['Survived'], 1).astype(float)) #features
y = np.array(train_data['Survived']) #target
print(X)
train_data.columns
```

I showed the X matrix. We had removed the 'Survived' section. The first column of the matrix is 'Pclass', the second column is 'Age', the third column is 'Fare'..

```python
#print(y)
```

We use the train_test_split function in order to make the split. The test_size=0.2 inside the function indicates the percentage of the data that should be held over for testing. The separation is 80/20(80% training set, 20% test set).

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Now we move on to the normalization section.
We use the StandardScaler() function of the Sklearn library to normalize the data.
Thanks to this function, we add bias to the data.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

## KMeans -  Clustering for Preprocessing

I will try to increase the accuracy value by adding a preprocessing on the best two models(rf and svm) in my previous classification project.<br>
I will use clustering for preprocessing step.

### Using SVM

Before applying preprocessing, let's apply the svm model on this data set and take a look at the accuracy value.

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC(C = 30, gamma = 0.01,random_state=0) #apply the model
svm.fit(X_train_sc, y_train) #training
```

```python
from sklearn.metrics import classification_report
prediction_svm = svm.predict(X_test_sc)
print(classification_report(y_test, prediction_svm))
```

This is first accuracy value.

```python
svm_acc1 = svm.score(X_test_sc,y_test)
print(svm_acc1) #first accuracy for svm
```

By using clustering, we can further increase this accuracy value.<br>
First, I gave the number of clusters randomly. I set the number of random clusters to 50.<br>
Using pipeline, I putted together different pieces.I applied clustering first, then applied the model. Finally, training was applied.

```python
pipeline_svm = Pipeline([
    ("kmeans", KMeans(n_clusters=50, random_state=1)),
    ("svm", SVC(C = 30, gamma = 0.01,random_state=0)),
])
pipeline_svm.fit(X_train_sc, y_train) #training
```

```python
svm_acc2 = pipeline_svm.score(X_test_sc,y_test)
print(svm_acc2) #second accuracy for svm
```

```python
1 - (1 - svm_acc2) / (1 - svm_acc1)
```

I reduced the error rate by over 5%. The accuracy value has increased, although not much.<br>
For better accuracy, we need to find the optimum number of clusters. For this, cross-validation(cv) was applied.<br>
Thanks to GridSearchCV, there is a exhaustive search over parameter values we have specified for the estimator.

```python
param_grid_svm = dict(kmeans__n_clusters=range(2, 100))
grid_clf_svm = GridSearchCV(pipeline_svm, param_grid_svm, cv=5, verbose=2)
grid_clf_svm.fit(X_train_sc, y_train) #training
```

This is the best number of clusters.

```python
grid_clf_svm.best_params_
```

```python
svm_acc3 = grid_clf_svm.score(X_test_sc, y_test)
print(svm_acc3) #third accuracy for svm
```

As you can see, a better accuracy value was obtained.

```python
1 - (1 - svm_acc3) / (1 - svm_acc2) # error rate
```

```python
svm_table = pd.DataFrame( 
    {  'Model' : ['SVM'], 
       'Accuracy' : [svm_acc1], 
       'Random clustering' : [svm_acc2],
       'Optimum clustering' : [svm_acc3]
     
    }) 

print(svm_table)
```

### Using Random Forest

Let's repeat the same process for random forest.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
```

n_estimators: number of trees in the forest.<br>
This parameter is used to determine the number of trees to be used in the process.

```python
random_forest = RandomForestClassifier(n_estimators=20, criterion='entropy', random_state=0) #apply the model
random_forest.fit(X_train_sc, y_train) #training
```

```python
prediction_rf = random_forest.predict(X_test_sc)
print(classification_report(y_test, prediction_rf))
```

Random forest has a higher accuracy than SVM.

```python
rf_acc1 = random_forest.score(X_test_sc, y_test)
print(rf_acc1) #first accuracy for rf
```

Firstly, I randomly determined the number of clusters -> n_clusters=20<br>
We applied the model again after clustering. 

```python
pipeline_rf = Pipeline([
    ("kmeans", KMeans(n_clusters=20, random_state=0)),
    ("random_forest", RandomForestClassifier(n_estimators=20, criterion='entropy', random_state=1)),
])
pipeline_rf.fit(X_train_sc, y_train) #training
```

Accuracy value increased slightly.

```python
rf_acc2 = pipeline_rf.score(X_test_sc,y_test)
print(rf_acc2) #second accuracy for rf
```

```python
1 - (1 - rf_acc2) / (1 - rf_acc1) #error rate
```

I reduced the error rate by over 3%. 

We can increase the accuracy value even more. We will use GridSearchCV () to find the optimum cluster value. I have explained in more detail in the SVM section.

```python
param_grid_rf = dict(kmeans__n_clusters=range(2, 100))
grid_clf_rf = GridSearchCV(pipeline_rf, param_grid_rf, cv=10, verbose=4)
grid_clf_rf.fit(X_train_sc, y_train) #training
```

This is the best number of clusters for random forest.

```python
grid_clf_rf.best_params_
```

This is the new accuracy value. As you can see, the accuracy value(rf_acc3) is higher than previous value(rf_acc2).

```python
rf_acc3 = grid_clf_rf.score(X_test_sc, y_test)
print(rf_acc3) #third accuracy for rf
```

```python
1 - (1 - rf_acc3) / (1 - rf_acc2) #error rate
```

```python
rf_table = pd.DataFrame( 
    {  'Model' : ['Random Forest'], 
       'Accuracy' : [rf_acc1], 
       'Random clustering' : [rf_acc2],
       'Optimum clustering' : [rf_acc3]
     
    }) 

print(rf_table)
```

## Comparison Table

I combined all the accuracy values of models in a table.

```python
table = pd.DataFrame( 
    {  'Model' : ['Random Forest', 'SVM'], 
       'Accuracy' : [rf_acc1, svm_acc1], 
       'Acc random clustering' : [rf_acc2, svm_acc2],
       'Acc optimum clustering' : [rf_acc3, svm_acc3]
     
    }) 

table = table.sort_values(by='Accuracy', ascending=False) #sorting
print(table)
```

A better accuracy was achieved thanks to clustering. Random Forest has better accuracy values than SVM.
